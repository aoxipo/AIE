```python
import cv2
import os

def batch_crop(directory, crop_coords, output_directory):
    for subdir, _, files in os.walk(directory):
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                file_path = os.path.join(subdir, file)
                img = cv2.imread(file_path)
                
                # 裁剪图像
                x, y, w, h = crop_coords
                cropped_img = img[y:y+h, x:x+w]
                
                # 创建输出目录
                output_subdir = subdir.replace(directory, output_directory)
                if not os.path.exists(output_subdir):
                    os.makedirs(output_subdir)
                
                # 保存裁剪后的图像
                output_path = os.path.join(output_subdir, f'{file}')
                print(output_path)
                cv2.imwrite(output_path, cropped_img)

```

```python
# 裁剪坐标: x, y, width, height
crop_coords = (0, 0, 768, 384)
# 指定的目录
directory = r'F:\Train\Train'

output_directory = r'F:\Train\ResizeTrain'

batch_crop(directory, crop_coords, output_directory)
```
