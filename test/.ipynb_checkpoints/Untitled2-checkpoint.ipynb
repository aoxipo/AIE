```python
import torch
import os
import numpy as np
import matplotlib.pyplot as plt
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
path = r"D:\dataset\eye\Train\Train/"
```

```python
Max = -1
Min = 999999
mark_path = ""
mark_path1 = ""
total_len = []
for folder in os.listdir(path):
    total_len.append(len(os.listdir( path + folder))  )
    if len(os.listdir( path + folder))> Max:
        Max = len(os.listdir( path + folder))
        mark_path = path + folder
    if len(os.listdir( path + folder))< Min and len(os.listdir( path + folder)) != 0:
        Min = len(os.listdir( path + folder))
        mark_path1 = path + folder
```
